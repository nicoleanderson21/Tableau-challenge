```python
jan2_df = jan_df.sample(n=100000)
feb2_df = feb_df.sample(n=100000)
march2_df = march_df.sample(n=100000)
april2_df = april_df.sample(n=100000)
may2_df = may_df.sample(n=100000)
june2_df = june_df.sample(n=100000)
july2_df = july_df.sample(n=100000)
aug2_df = aug_df.sample(n=100000)
sept2_df = sept_df.sample(n=100000)
oct2_df = oct_df.sample(n=100000)
nov2_df = nov_df.sample(n=100000)
dec2_df = dec_df.sample(n=100000)
```

```python
combo_df = pd.concat([jan2_df, feb2_df, march2_df, april2_df, may2_df, june2_df, july2_df, aug2_df, sept2_df, oct2_df, nov2_df, dec2_df], axis=0)
len(combo_df)
```

```python
combo_df['started_at'] = pd.to_datetime(combo_df['started_at'], format='%Y-%m-%d %H:%M:%S')
combo_df['ended_at'] = pd.to_datetime(combo_df['ended_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
combo_df['start_date'] = pd.to_datetime(combo_df['started_at']).dt.date
combo_df['start_time'] = pd.to_datetime(combo_df['started_at']).dt.time
combo_df['end_date'] = pd.to_datetime(combo_df['ended_at']).dt.date
combo_df['end_time'] = pd.to_datetime(combo_df['ended_at']).dt.time
combo_df.head()
```

```python
combo_df["trip_duration"] = (pd.to_datetime(combo_df["end_time"].astype(str), format='%H:%M:%S') - pd.to_datetime(combo_df["start_time"].astype(str), format='%H:%M:%S'))
combo_df.head()
```

```python
combo_df['trip_duration'] = combo_df['trip_duration']/pd.Timedelta(minutes=1)
combo_df.head()
```

```python
def haversine(lat1, lon1, lat2, lon2):
    
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

   
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371 
    return c * r
```

```python
combo_df['Distance_traveled'] = haversine(combo_df['start_lat'].shift(), combo_df['start_lng'].shift(), combo_df['end_lat'], combo_df['end_lng'])
combo_df.head()
```
